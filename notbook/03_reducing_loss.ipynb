# 降低损失 (Reducing Loss)：梯度下降法


[Link to gradient-descent](https://developers.google.com/machine-learning/crash-course/reducing-loss/gradient-descent)

## 目的：
通过用梯度下降来解决一维的问题来理解梯度下降法，这样可以将梯度下降同样推广到多维的情况。

## 问题：
假设有100对观测到的数值(x, y)（由 y=b+w*x，w=5, b=4生成。具体的数值事先不知道），如何通过梯度下降法来获取b和w的值？



## 生成100组数据

```python
import numpy as np
# y = b+w*x
w = 5
b = 4

x = np.arange(100)
y = b + w * x
```

## 使用梯度下降法

```python
# loss = (y-(b1+w1*x))2 # 损失函数
# g_w1 = 2*w1*(y-b1-w1*x) # 对w的偏导数
# g_b1 = 2*(y-b1-w1*x) # 对b的偏导数

learing_rate = 0.0001
w1 = 0
b1 = 0

def g_w1(x, y, b1, w1):
    return -2*x*(y-b1-w1*x)

def g_b1(x, y, b1, w1):
    return -2*(y-b1-w1*x)

print(f"w1: {w1}, b1: {b1}")


```

```python
for i in range(1000):
    for x1, y1 in zip(x, y):
        w1 = w1 - g_w1(x1, y1, b1, w1) * learing_rate
        b1 = b1 - g_b1(x1, y1, b1, w1) * learing_rate
    print(f"w1: {w1}, b1: {b1}")
```
